```python
# Extracting year from release date
movies_data['Release Year'] = movies_data['Release Date'].dt.year
```

```python
# movies release per year.
movies_per_year = movies_data['Release Year'].value_counts().sort_index()
movies_per_year
```

```python
# line chart showing the number of movies release in each year.

plt.figure(figsize=(12, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-', color='b')
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Runtime movie distribution

plt.figure(figsize=(10, 6))
sns.histplot(movies_data['Runtime'], bins=30, kde=True)
plt.title('Distribution of Movie Runtime')
plt.xlabel('Runtime')
plt.ylabel('Frequency')
plt.show() 
```

```python
revenue=movies_data[['Title','Revenue','Release Year']]
revenue = revenue.sort_values(by='Revenue', ascending=False)
```

```python
revenue_by_year = movies_data.groupby('Release Year')['Revenue'].sum().sort_values(ascending=False)
revenue_by_year
```

```python
highest_revenue = revenue_by_year.max()
highest_revenue
```

```python
plt.figure(figsize=(12, 6))
plt.plot(revenue_by_year.index, revenue_by_year.values, marker='o', linestyle='-', color='b')
plt.title('Total Revenue by Year')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```
